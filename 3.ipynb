```python
import pandas as pd
import matplotlib.pyplot as plt
# Загрузка данных
data = pd.read_csv('D:\GB Codes\kc_house_data.csv')
plt.figure(figsize=(12, 6))
plt.scatter(data['bedrooms'], data['price'], alpha=0.5,
color='blue')
plt.title('Цена недвижимости в зависимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена (USD)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Загрузка данных
data = pd.read_csv('D:\GB Codes\kc_house_data.csv')
plt.figure(figsize=(12, 6))
plt.boxplot([data[data['condition'] == i]['price'] for i in
sorted(data['condition'].unique())],
labels=sorted(data['condition'].unique()))
plt.title('Цена недвижимости в зависимости от состояния')
plt.xlabel('Состояние дома')
plt.ylabel('Цена (USD)')
plt.grid(True)
plt.show()
```
