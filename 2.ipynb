```python
import pandas as pd
import matplotlib.pyplot as plt
# Загрузка данных
data = pd.read_csv('D:\GB Codes\kc_house_data.csv')
# Построение графика распределения домов по наличию вида на набережную
plt.figure(figsize=(8, 5))
data['waterfront'].value_counts().plot(kind='bar', color=['skyblue',
'salmon'])
plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.xticks([0, 1], ['Нет', 'Да'], rotation=0)
plt.grid(axis='y')
plt.show()
```
